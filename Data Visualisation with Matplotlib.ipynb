## Subplots
- plt.subplot(n,m,i) or plt.subplot(nmi) 
### n represent the row, m represent column and i represent the position of the chart




```python
plt.figure(figsize=(6,4))
#plot 1
x=np.array([0,1,2,3])

y=np.array([3,8,1,10])
plt.subplot(1,3,1)  #plot position(1) is numbered per column
plt.plot(x,y)
plt.title("Plot 1")

# #plot2
b=np.array([0,1,2,3])
c=np.array([10,20,30,40])
plt.subplot(1,3,2)
plt.plot(b,c)
plt.title("Plot 2")
 
# #plot3
d=np.array([0,1,2,3])
e=np.array([10,20,30,40])
plt.subplot(1,3,3)
plt.plot(d,e)
plt.title("Plot 3")

plt.tight_layout()
plt.show()
```

## Pie Chart
- Pie chart represents a part-to-whole relationship
- Each pie is composed of three components; area,central angle and arc lenght.


```python
#Define the figure size
plt.figure(figsize=(7,7))

marks=[25,30,45,10]

#labels of the pie
Names=['Ama','Akosua','Emma','Harley']
plt.pie(marks,labels=Names)
plt.legend()  #plt.legend can only be used when labels property have been assigned for each plot
plt.show()


```

## Bar Chart

- A bar chart is used to represent comparative data

```python
x=[1,2,3,4,5]
y=[2,4,11,8,10]

plt.bar(x,y)
plt.show()
```

## Scatter Plots

Scatter plots are similar to regular plots but they show individuals points instead of values connected in series.

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y, marker='*',color='g')
plt.title('Our sine graph')
plt.xlabel('x axis')
plt.ylabel('sinX')
plt.grid()
plt.show()
```

## Histogram
Histogram is a bar representation of data that varies over a range. It plots the height of the data belonging to a range on the y-axis and plots the range along the x-axis. It use a bar representation to show the data belonging to each range.
It divides the ranges into bins.

```python
from random import randint
randint(15,80)
#help (randint)
```

```python
range(100)
```

```python
from random import randint
x=[randint(15,80) for i in range(100)]  #list comprehension, randint[15,80] inclusive

y=[20,30,40,50,60,70,80]

plt.hist(x, bins=y,color='m',ec='red')  #ec, edge color
plt.title('hist of ages')
plt.xlabel('age group')
plt.ylabel('no of students')
plt.show()
```

## Grouped Bar Chart

```python
plt.figure(figsize=(5,5))

labels=['Mon','Tue','Wed','Thu','Fri']
mobile_sales=[20,34,30,35,27]
pc_sales=[25,32,34,20,25]

x=np.arange(len(labels)) #the label location
width=0.35 #the width of the bars
a=plt.bar(x-width/2,mobile_sales,width,color='m', label='Mobile')
b=plt.bar(x+width/2,pc_sales,width,color='k', label='PC')

plt.ylabel('Total sales')
plt.title('Total sales by mobile and pc')
plt.xticks(x,labels) #gets location for the labels on x-axis
plt.legend()
#plt.bar_label(a) #bar label adds labels to bars
#plt.bar_label(b)
#plt.savefig('groupedbar1.png')
plt.show()
```

## Stacked Bar Chart

Some times if you want to see how a larger category is divided into smaller and what the relationship
<br> of each part has on the total amount 

```python
#data for 2016 olympics
plt.figure(figsize=(5,7))

x=['USA','GBR','China','Russia','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
w=0.5 #default is 0.8
c='m'
cr='y'
plt.bar(x,gold,w,color=cr)  #gold represented by colour blue
plt.bar(x,silver,w, bottom=gold,color=c)   #silver represented by color orange

plt.show()
```
